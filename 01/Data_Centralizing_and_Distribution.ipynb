```python
import numpy as np
import math
from scipy import stats
import statistics as s
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# MEAN
print("Mean:")
apel = [100, 200, 150, 100, 120, 80, 90, 160, 110, 170]
avg_apel = sum(apel) / len(apel)
print(avg_apel)

# Median
print("\nMedian:")
def bubblesort(elements):
    for n in range(len(elements)-1, 0, -1):
        for i in range(n):
            if elements[i] > elements[i + 1]:
                elements[i], elements[i + 1] = elements[i + 1], elements[i]
    return elements

def calc_median(elements):
    elements = bubblesort(elements)
    len_elements = len(elements)
    if len_elements % 2 == 0:
        center = len_elements // 2
        return (elements[center-1] + elements[center]) / 2
    else:
        return elements[math.floor(len_elements / 2)]

apel = [100, 200, 150, 100, 120, 80, 90, 160, 110, 170]
median = calc_median(apel)
print(median)

# Mode
print("\nMode:")
nilai = [70, 80, 70, 70, 90, 100]

value_mode = stats.mode(nilai)
modus = s.mode(nilai)
print(modus)
print(value_mode)
print(value_mode.mode)

# Interquartile Range
print("\nInterquartile Range:")
data_baru = np.array([23, 56, 45, 65, 59, 55, 62, 54, 85, 25, 55])
print(pd.DataFrame(data_baru).describe())

# Normalization 1
print("\nNormalization 1:")
data_raw = np.array([
    [2, 3, 7, 30],
    [9, 4, 6, 1],
    [8, 15, 2, 40],
    [20, 10, 2, 6]
])
print(data_raw)

# Normalization 2
print("\nNormalization 2:")
data_raw = np.array([
    [2, 3, 7, 30],
    [9, 4, 6, 1],
    [8, 15, 2, 40],
    [20, 10, 2, 6]
])

scaler = MinMaxScaler()
data_scale = scaler.fit_transform(data_raw)
print(data_scale)

# Standardization
print("\nStandardization:")
data_raw = np.array([
    [2, 3, 7, 30],
    [9, 4, 6, 1],
    [8, 15, 2, 40],
    [20, 10, 2, 6]
])

scaler = StandardScaler()
data_stand = scaler.fit_transform(data_raw)
print(data_stand)
```
