**Find the mean, median, and mode!**


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# List of variables
variables = ['age', 'amount_of_loan', 'period_in_year', 'credits_per_month',
             'id_collector', 'guarantee', 'nominative_account', 'main_arrears', 'interest_arrears']
```

```python
# Calculate and print mean, median, and mode for each variable
for variable in variables:
    print(f"\nAnalysis for {variable}:")
    print(f"Mean: {df[variable].mean()}")
    print(f"Median: {df[variable].median()}")
    print(f"Mode: {df[variable].mode().iloc[0]}")
```

**Determine the Q1, Q2, Q3, and the IQR! Also variance and its standard deviation!**


```python
# Calculate and print quartiles, IQR, variance, and standard deviation for each variable
for variable in variables:
    print(f"\nAnalysis for {variable}:")
    print(f"Q1 (25th percentile): {df[variable].quantile(0.25)}")
    print(f"Q2 (50th percentile): {df[variable].quantile(0.5)}")
    print(f"Q3 (75th percentile): {df[variable].quantile(0.75)}")
    print(f"IQR (Interquartile Range): {df[variable].quantile(0.75) - df[variable].quantile(0.25)}")
    print(f"Variance: {df[variable].var()}")
    print(f"Standard Deviation: {df[variable].std()}")
```

**Do you find some outliers? Explain it!**

```python
# Check for outliers and print the results for each variable
for variable in variables:
    lower_bound = df[variable].quantile(0.25) - 1.5 * (df[variable].quantile(0.75) - df[variable].quantile(0.25))
    upper_bound = df[variable].quantile(0.75) + 1.5 * (df[variable].quantile(0.75) - df[variable].quantile(0.25))
    outliers = df[(df[variable] < lower_bound) | (df[variable] > upper_bound)]
    print(f"\nOutliers for {variable}:")
    print(f"Number of outliers: {len(outliers)}")
    print("Outliers:", outliers)
```


**Make a histogram based on your work!**

```python
# Create histograms for each variable
for variable in variables:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[variable], bins='auto', kde=True, color='#0504aa', alpha=0.7)
    plt.title(f'Histogram of {variable}')
    plt.xlabel(variable)
    plt.ylabel('Frequency')
    plt.grid(axis='y', alpha=0.75)
    plt.show()
```
